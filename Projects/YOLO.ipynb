#### YOLO Models
* The YOLO model first described by Joseph Redmon, et al, in the 2015 paper titled "You Only Look Once: unified, Real-time object detection.
* Note that Ros Girshick, developer of R-CNN, was also an author and contributor to his work, then at Facebook AI Research.
* The approach involves a single nueral network trained end to end that takes a photograph as input and predicts bounding boxes and class labels for each bounding box directly.
* The rechnique offers lower predictive accuracy (e.g. more localization errors), although operates at 45 frames per second and up to 155 frames per second for a speed-optimized version of the model.

```python
import torch
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
# Define the function for loading the image
def load_image(image):
    image = Image.open(image)
    return image
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
# Define the predict function
def predict(image):
  # Load the iamge
  image = load_image(image)

  # run the model inference
  results = model(image)

  # print the detected objects
  print('Detected Objects: ', results.names)

  results.show()

  results.save()
```
