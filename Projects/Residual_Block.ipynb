#### Residual Block
* The ResNet architecture introduces the simple concept of adding an intermediate input tot the output of a series of convolution blocks.

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.applications.resnet50 import decode_predictions
```

```python
image = load_img(r'C:\Users\DAI.STUDENTSDC\Desktop\6.NLP\Projects\images\adolf.webp', 
                 target_size=(224, 224))
```

```python
image
```

```python
type(image)
```

```python
# convert the image to array
img_to_array(image)
```

```python
# reshape the image
image = img_to_array(image)

image = image.reshape(1, 224, 224, 3)
```

```python
# Preprocess the image for vgg16 model

image = preprocess_input(image)
```

```python
model = ResNet50()
```

```python
result = model.predict(image)
```

```python
label = decode_predictions(result)
```
