#### Thresholding
* Thresholding is a fundamental technique in computer vision used to segment an image into foreground and background objects.
* It works by classifying each pixedl in the image as either belonging to the object of interest(foreground) or the background based on their intensity values.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/sachin.jpg', 0)
```

```python
binary_image = img > 127
```

```python
binary_image
```

```python
plt.figure(figsize=(16,9))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.title('Binary Image')
plt.imshow(binary_image, cmap='gray')
```

```python
lenna = cv2.imread('images/lenna.png', 0)

binary_lenna = lenna > 127

plt.figure(figsize=(16,9))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(lenna, cmap='gray')

plt.subplot(1,2,2)
plt.title('Binary Image')
plt.imshow(binary_lenna, cmap='gray')
```

```python
lenna = cv2.imread('images/lenna.png', 0)

binary_lenna = lenna > 127

new_lenna  = binary_lenna.astype(np.uint8) * 255 

plt.figure(figsize=(16,9))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(lenna, cmap='gray')

plt.subplot(1,2,2)
plt.title('Binary Image')
plt.imshow(new_lenna, cmap='gray')
```

```python
## The above code can be re-written using the built-in functions

lenna = cv2.imread('images/lenna.png', 0)

thresh, new_lenna = cv2.threshold(lenna,
                                  127,
                                  255,
                                  cv2.THRESH_BINARY)  # THRESH_BINARY is the built-in function

plt.figure(figsize=(16,9))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(lenna, cmap='gray')

plt.subplot(1,2,2)
plt.title('Binary Thresholded Image')
plt.imshow(new_lenna, cmap='gray')
```

#### Inverse Binary Thresholding

```python
lenna = cv2.imread('images/lenna.png', 0)

binary_lenna = lenna < 127 # lenna < 127 inverts the image.

new_lenna  = binary_lenna.astype(np.uint8) * 255 

plt.figure(figsize=(16,9))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(lenna, cmap='gray')

plt.subplot(1,2,2)
plt.title('Binary Image')
plt.imshow(new_lenna, cmap='gray')
```

```python
## The above code can be re-written using the built-in functions

lenna = cv2.imread('images/lenna.png', 0)

thresh, new_lenna = cv2.threshold(lenna,
                                  127,
                                  255,
                                  cv2.THRESH_BINARY_INV) # the function THRESH_BINARY_INV inverts the thresholding

plt.figure(figsize=(16,9))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(lenna, cmap='gray')

plt.subplot(1,2,2)
plt.title('Binary Thresholded Image')
plt.imshow(new_lenna, cmap='gray')
```

#### Otsu's thresholding
* Otsu's thresholding is a poupular and effective automatic thresholding technique used in image processing to separate an image into foreground and background based on pixel intensity.
* It aims to find the threshold value that maximizes the between-class variance, essentially creating the best possible separation between the two classes (foreground and background).

```python
## The above code can be re-written using the built-in functions

lenna = cv2.imread('images/lenna.png', 0)

thresh_bin, lenna_bin = cv2.threshold(lenna,
                                  100,
                                  255,
                                  cv2.THRESH_BINARY) # the function THRESH_OTSU inverts the thresholding


thresh_otsu, lenna_otsu = cv2.threshold(lenna,
                                  100,
                                  255,
                                  cv2.THRESH_OTSU) # the function THRESH_OTSU inverts the thresholding


plt.figure(figsize=(16,9))

plt.subplot(1,3,1)
plt.title('Original Image')
plt.imshow(lenna, cmap='gray')

plt.subplot(1,3,2)
plt.title('Binary Thresholded Image')
plt.imshow(lenna_bin, cmap='gray')

plt.subplot(1,3,3)
plt.title('Otsu Thresholded Image')
plt.imshow(lenna_otsu, cmap='gray')
```
