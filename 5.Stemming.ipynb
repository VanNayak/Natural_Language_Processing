Stemming is the process of reducing a word to its stem that affixes to the suffixes and prefixes or to the roots of words known as "lemmas".

```python
import nltk
```

```python
words1 = 'managing', 'manage', 'management', 'managerial', 'managed'
words2 = 'worked', 'workable', 'working', 'works'
words3 = 'big', 'biggest', 'bigger'
words4 = 'go', 'gone', 'went', 'going'
words5 = 'association', 'associated', 'associate'
```

### Porter Stemmer

```python
ps.stem(words1[2])
```

```python
text = 'Hello friends! How are you? I like Python programming'
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
```

```python
nltk.download('punkt_tab')

tokens1 = [word for word in word_tokenize(text) if word not in string.punctuation]
```

```python
tokens1
```

```python
nltk.download('stopwords')

swords = stopwords.words('english')
tokens2 = [word for word in tokens1 if word.lower() not in swords]
```

```python
tokens2
```

```python
# Stemming

swords = stopwords.words('english')
tokens3 = [ps.stem(word) for word in tokens1 if word.lower() not in swords]
```

```python
tokens3
```
