```python
from nltk.corpus import stopwords
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
text ="Today is Friday! The 'best thing' to do today is to sit at home, watch a movie and and sip ...may be 'coffee'. "
```

```python
from nltk.tokenize import word_tokenize
import string
```

```python
tokens = [word for word in word_tokenize(text) if word.isalpha() or word.isdigit()]
```

```python
 string.punctuation
```

```python
' '.join(tokens)
```

```python
# Stopwords Removal
swords = stopwords.words('english')
```

```python
text = ' '.join(tokens)
```

```python
swords
```

```python
words = [word for word in word_tokenize(text) if word.lower() not in swords]
```
